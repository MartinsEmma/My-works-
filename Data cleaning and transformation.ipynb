## REPLACING EMPTY VALUES

```python
##WHILE DEALING WITH NUMERICAL CLOLUMNS(COLUMNS THAAT CONTAIN INT OR FLOATS OR NUMLIKE FIGS), WE REPLACE WITHE THE MEAN OR MEDIAN
##IN DEALING WITH CATEGORICAL COLUMNS WE(COLUMNS THAT CONTAINS STR), WE REPLACE WITH THE MODE
```

```python
null_data
```

```python
##REPLACING MISSING VALUES WITH THE MEAN 
df_mean = df['Calories'].mean()
df_mean
```

```python
## CREATING A CHECKPOING 
df2 = df.copy()
```

```python
df2['Calories'] = df['Calories'].fillna(df_mean)
df2.info()
```

```python
df2
```

```python
##REPLACING MISSING VALUES WITH THE MEDIAN
df_median = df['Calories'].median()
df_median
```

```python
df3 = df2.copy()
```

```python
df3['Calories'] = df2['Calories'].fillna(df_median)
df3.info()
```

```python
df3
```
