##  Hypothesis Analysis

### Interest rate is Varied for different loan amounts 
* H0: Interest rate is not varied for different loan amounts (no significant difference)
* H1: Interest rate is varied for different loan amounts (significant difference)

```python
from scipy.stats import ttest_ind

#Defining high amount and low loan amount groups
high_amount = data[data['Amount_Requested']  > data['Amount_Requested'].median()]
low_amount = data[data['Amount_Requested']  <= data['Amount_Requested'].median()]

#Performing t-test 
t_stat, p_val = ttest_ind(high_amount['Interest_Rate'], low_amount['Interest_Rate'])
print(f"T-statistic: {t_stat}, p-value: {p_val}\n")

print('At 95% confidence level\n')

#Interpertation
if p_val < 0.05:    
    print("Rejecting the null hypothesis: Interest rate is significantly different for high vs low amounts.\n")
    print("Accepting the alternative hypothesis: Interest rates vary depending on the loan amount.")
else:
    print("Fail to reject the null hypothesis: No significant difference in interest rate for high vs low amounts.")
    print("This suggests that interest rates do not vary significantly based on the loan amount.")
```

## Loan length is directly effecting interest rate.
* H0: Loan length is not affecting interest rate
* H1: Loan length is directly affecting rate

```python
data.columns
```

```python
from scipy.stats import f_oneway 

#Grouping data by loan length and collect interest rates 
loan_length_36 = data[data['Loan_Length'] == 36]['Interest_Rate']
loan_length_60 = data[data['Loan_Length'] == 60]['Interest_Rate']

#Performing ANOVA
f_stat, p_val = f_oneway(loan_length_36, loan_length_60)
print(f"F-statistic: {f_stat}, p-value: {p_val}\n")

print('At 95% confidence level\n')

#Interpretation of result 
if p_val < 0.05: 
    print("Rejecting the null hypothesis: Loan length significantly affects the interest rate.")
    print("Accepting the alternative hypothesis: Loan length is directly affecting the interest rate.")
else:
    print("Fail to reject the null hypothesis: Loan length does not significantly affect the interest  rate.")
    print("This suggests that loan length is not directly affecting the interest rate.")
```

# Interest Rate varies for different purpose of loans 
### *H0: Interest rate does not vary for different purpose of loans
### *H1: Interest rate varies for different purpose of loans

```python
data.Loan_Purpose.nunique()
```

```python
g1 = data.loc[data['Loan_Purpose'] == 'car']['Interest_Rate']
g2 = data.loc[data['Loan_Purpose'] == 'credit_card']['Interest_Rate']
g3 = data.loc[data['Loan_Purpose'] == 'debt_consolidation']['Interest_Rate']
g4 = data.loc[data['Loan_Purpose'] == 'educational']['Interest_Rate']
g5 = data.loc[data['Loan_Purpose'] == 'home_improvement']['Interest_Rate']
g6 = data.loc[data['Loan_Purpose'] == 'house']['Interest_Rate']
g7 = data.loc[data['Loan_Purpose'] == 'major_purchase']['Interest_Rate']
g8 = data.loc[data['Loan_Purpose'] == 'medical']['Interest_Rate']
g9 = data.loc[data['Loan_Purpose'] == 'moving']['Interest_Rate']
g10 = data.loc[data['Loan_Purpose'] == 'other']['Interest_Rate']   
g11 = data.loc[data['Loan_Purpose'] == 'renewable_energy']['Interest_Rate']
g12 = data.loc[data['Loan_Purpose'] == 'small_business']['Interest_Rate']
g13 = data.loc[data['Loan_Purpose'] == 'vacation']['Interest_Rate']
g14 = data.loc[data['Loan_Purpose'] == 'wedding']['Interest_Rate']
```

```python
from scipy.stats import f_oneway

#Performing ANOVA 
f_stat, p_val = f_oneway(g1, g2, g3, g4, g5, g6, g7, g8, g9, g10, g11, g12, g13, g14)

print(f"F-statistic : {f_stat}, p-value: {p_val}\n")

#Interpretation
if p_val < 0.05:
    print("Reject the null hypothesis: Interest rate varies for different purposes of loans.")
else:
    print("Fail to reject the null hypothesis: Interest rate does not vary for different purposes of loans.")
```

# Checking the relationship between FICO and Score Range and Home Ownership
### *H0: There is no relationship between FICO Scores and Home Ownership
### *H1: There is a relationship betwee FICO and Home Ownership

```python
data['Home_Ownership'].value_counts()
```

```python
from scipy import stats 

#Creating the crosstab
Cross_Table = pd.crosstab(data['FICO_Range'], data['Home_Ownership'], margins=True)

#Performing Chi-square Test
chi2, p_val, dof, expected = stats.chi2_contingency(Cross_Table)

#Printing Results and Interpretations
print("Chi-Square Test Resullts\n")
print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"p-value: {p_val:.4f}")
print("\nInterpretation:")
if p_val < 0.05: 
    print("Reject H0: There is a relationship between FICO and Home Ownership.")
else:
    print("Fail to reject H0: There is no relationship between FICO and Home Ownership.")
```
