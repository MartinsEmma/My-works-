```python
# Drop rows with missing values in either X or y
X = x.dropna()
y = y[X.index]  # Ensure y aligns with X
```

## Specify and Fit Model# 
Create a DecisionTreeRegressor and save it iowa_model. Ensure you've done the relevant import from sklearn to run this command.# 

Then fit the model you just created using the data in X and y that you saved above.

```python
# Step 1: Import the model class
from sklearn.tree import DecisionTreeRegressor

# Step 2: Create the model
ML_model = DecisionTreeRegressor(random_state=1)

# Step 3: Fit the model
ML_model.fit(X, y)
```

## Make Predictions### 
Make predictions with the model's predict command using X as the data. Save the results to a variable called predictions.

```python
# Make predictions using the trained model
predictions = ML_model.predict(X)

# Display the first few predictions
print(predictions[:5])
```

```python
# Create a new DataFrame with the features
results = X.copy()

# Add a new column for the predictions
results['PredictedPrice'] = predictions

# Display the first few rows
print(results.head())
```

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = predictions
mean_absolute_error(y, predicted_home_prices)
```

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define model
ML_model = DecisionTreeRegressor()
# Fit model
ML_model.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = ML_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```
